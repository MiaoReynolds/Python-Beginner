# Import the library

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

plot something

```python
plt.scatter(1,1);
```

define the plot figure size

```python
plt.figure(figsize=(20,10));
```

# Subplot

```python
for i in range (1, 5):      # for python, (1,5) the 1 is included but 5 is excluded
    plt.subplot(2, 2, i)    # i is 1, 2, 3, 4
    plt.text(0.5,0.5, str((2, 2, i)),horizontalalignment='center', verticalalignment='center', fontsize = 10)
                            # 0.5,0.5 is the location of the text head
                            # str(x,y,z) is the contents of the text
                            # hori. and verti. are the alighment of the text
                            # fontsize...
    plt.tight_layout()
```

```python
for i in range (1, 5):      # for python, (1,5) the 1 is included but 5 is excluded
    plt.subplot(2, 2, i)    # i is 1, 2, 3, 4
    plt.text(0.5,0.5, str((2, 2, i)), fontsize = 10, color='red')
                            # 0.5,0.5 is the location of the text head
                            # str(x,y,z) is the contents of the text
                            # hori. and verti. are the alighment of the text, without it, the text head will start from 0.5,0.5
                                # see red text in below
                            # fontsize...
    plt.tight_layout()
```

# Now we change some line style

scatter will draw points
plot will draw straight line

```python
x=[1,2,3,4]
y=[2,4,6,7]
plt.subplot(1, 2, 1) 
plt.scatter(x, y, marker = 'o', color = 'red' , linewidth = 2, linestyle ='-');
                                # there are many marker choice
                                # there are many color
                                # scatter will not show line style even stated
plt.subplot(1, 2, 2) 
plt.plot(x, y, color = 'red' , linewidth = 2, linestyle ='-');
                                # plot will draw line and line style
plt.tight_layout()
```

# Now we put some legend

```python
plt.scatter(x, y, marker = 'o', color = 'red');
a=[1,2,3,4]
b=[2.5,4.3,5.5,7.9]
                                # also, we want more scatter
                                # a and b are new scatter data
plt.scatter(a,b,marker='*',color='green')
                                # tight_layout is to fit the plot to the full size
plt.ylabel('Income', fontsize = 15) #for y label
plt.xlabel('Age', fontsize = 15) #for x label
my_labels = ['happy','green-dot']
                                # we want to label the lines/scatter
plt.legend(labels = my_labels, loc = 'lower right', fontsize = 20);
plt.tight_layout()
```

# Add annotation

```python
plt.figure(figsize=(10,10));
plt.plot(x, y, marker = 'o', color = 'red');
plt.plot(a,b,marker='*',color='green')
plt.annotate('here!', xytext=(3, 5), xy = (2, 4), arrowprops=dict(facecolor='black', shrink=0.05))
                                # the annotation xy must on data point or line, or it will not show
plt.tight_layout()
plt.show()
```
